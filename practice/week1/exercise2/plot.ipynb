```python
import matplotlib.pyplot as plt

# Read data from the text file
with open('eigenvals.txt', 'r') as file:
    data = [float(line.strip()) for line in file]

# Generate x values from 1 to 10
x_values = list(range(1, 11))

# Plot the graph
plt.plot(x_values, data, marker='o', linestyle='-')
plt.title('Data Plot')
plt.xlabel('Dimension')
plt.ylabel('Largest eigenvalue')
plt.xlim(1, 10)
plt.ylim(0, max(data) + 0.5) 
plt.grid(True)
plt.show()
```
